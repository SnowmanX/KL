```python
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt
# 随机生成两个离散型分布
x = [np.random.randint(1, 11) for i in range(10)]
# print(x)
print(np.sum(x))
px = x / np.sum(x)
# print(px)
y = [np.random.randint(1, 11) for i in range(10)]
# print(y)
print(np.sum(y))
py = y / np.sum(y)
# print(py)

# 利用scipy API进行计算
# scipy计算函数可以处理非归一化情况，因此这里使用
# scipy.stats.entropy(x, y)或scipy.stats.entropy(px, py)均可
KL = scipy.stats.entropy(x, y) 
print(KL)

# 编程实现
KL = 0.0
for i in range(10):
    KL += px[i] * np.log(px[i] / py[i])
    # print(str(px[i]) + ' ' + str(py[i]) + ' ' + str(px[i] * np.log(px[i] / py[i])))

print(KL)

plt.subplot(121)
plt.hist(x)
plt.subplot(122)
plt.hist(y)
```

```python
%matplotlib inline
from hpelm.elm import ELM
import scipy.io as sio
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import KFold
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import GroupKFold
from sklearn.preprocessing import MinMaxScaler
import itertools
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import OneHotEncoder
from functools import reduce
from scipy import signal
from sklearn.metrics import confusion_matrix
import datetime
import h5py

from sklearn.ensemble import RandomForestClassifier
import numpy as np
import matplotlib.pyplot as plt 
import mpl_toolkits.mplot3d
from sklearn.mixture import GaussianMixture

FILE = 'data_raw'


def load_mat(file):
    data = sio.loadmat(file)
    data_idx = list(data.keys())
    data = data[data_idx[-1]]
    data = rm_zeros_row(data)
    return data


def load_feather(file):
    df = pd.read_feather(file)
    # df_shuffle = df.sample(frac=1, random_state=666)
    return df.values


def rm_zeros_row(data):
    zeros_cond = (data[:, 3:] == 0)
    zeros_row_index = (zeros_cond.sum(axis=1) == 270)
    data_wto = data[~zeros_row_index]
    return data_wto


def validate(x, validation_x):
    pass


def reprocess_phase(phase_data):
    phase_matrix = np.zeros([int(np.max(phase_data[:, 0])), 273])
    for i in range(int(np.max(phase_data[:, 0]))):
        phase_matrix[i, :] = np.sum(phase_data[phase_data[:, 0] == (i + 1), :], axis=0) / \
            (phase_data[:, 0] == (i + 1)).sum()
    return phase_matrix


def sliding_windows(data, width=10, stride=10):
    start = 0
    end = 0
    slide_times = int(8000 / stride)
    num = int(np.max(data[:, 0]))
    sliding_data = np.zeros([num * slide_times, 273])
    real_end = 0
    for i in range(int(np.max(data[:, 0]))):
        # end = (data[:, 0] == i + 1).sum()
        for j in range(slide_times):
            end = start + width
            sliding_data[i * 9 + j,
                         :] = np.sum(data[start:end, :], axis=0) / width
            start = stride + start
        end = (data[:, 0] == i + 1).sum()
        start = end + start
    return sliding_data


def accuracy(pred, target):
    return np.sum(np.sqrt(np.sum(np.square(pred - target), axis=1))) / target.shape[0]


def test_validation(train_x_trans, train_target, test_x_trans, test_target):
    accuracy_train = np.zeros([330])
    accuracy_validation = np.zeros([330])
    for i in range(1, accuracy_train.shape[0] + 1):
        accuracy_train_acc = 0
        accuracy_validation_acc = 0
        for j in range(200):
            elm = ELM(train_x_trans.shape[1], train_target.shape[1])
            elm.add_neurons(i, 'sigm')
            err = elm.train(train_x_trans, train_target, 'v')
            train_pred = elm.predict(train_x_trans)
            validation_pred = elm.predict(test_x_trans)
            accuracy_train_acc = accuracy_train_acc + np.sum(np.sqrt(
                np.sum(np.square(train_pred - train_target), axis=1))) / train_target.shape[0]
            accuracy_validation_acc = accuracy_validation_acc + np.sum(np.sqrt(
                np.sum(np.square(validation_pred - test_target), axis=1))) / test_target.shape[0]
        accuracy_train[i - 1] = accuracy_train_acc / 200
        accuracy_validation[i - 1] = accuracy_validation_acc / 200
        print(accuracy_train[i - 1])
        print(accuracy_validation[i - 1])
        print('count:', i)


def butterworh_filter(data):
    fn = 1000
    fs = 5 / (1000 / 2)
    b, a = signal.butter(3, 0.08, 'low')
    sf = signal.filtfilt(b, a, data.T)
    df = pd.DataFrame(sf)
    df.columns = [str(i) for i in range(data.shape[0])]
    df.to_csv(datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S'))
    return sf


def downsampling(data, stride=10):
    data_sampling = np.zeros([np.floor(len(data) / 10), data.shape[1]])
    count = 0
    for i in range(len(data)):
        if (i + 1) % stride == 0:
            data_sampling[count, :] = data[i, :]
            count = count + 1
    return data_sampling


def reduce_sum(data, stride):
    from functools import reduce
    data_point_num = int(len(data) / stride)
    return np.vstack([np.sum(data[i * stride:(i + 1) * stride, :], axis=0) / stride for i in range(data_point_num)])


def weight_computing(list):
    return np.exp(list) / np.sum(np.exp(list))


# perform data and testing data generation
def data_re(data, index):
    data_label = data[:, 0]
    test_index = reduce(lambda x, y: x | y, [data_label == i for i in index])
    data_tmp = data[~test_index, :]
    data_tmp_test = data[test_index, :]
    return data_tmp, data_tmp_test


def best_elm_accuracy(train_x_trans, train_target, test_x_trans, times, optimal_neurons):
    train_accuracy_list = []
    test_accuracy_list = []
    for i in range(times):
        best_elm = ELM(train_x_trans.shape[1], train_target.shape[1])
        best_elm.add_neurons(optimal_neurons)
        best_elm.train(train_x_trans, train_target)


def softmax(x):
    return (np.exp(x).T / np.exp(x).sum(axis=1)).T, (np.exp(x).T / np.exp(x).sum(axis=1)).T.argmax(axis=1).reshape([-1, 1])


def score(pred, target):
    return (pred == target).sum() / len(target)


def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
#         print("Normalized confusion matrix")
#     else:
#         print('Confusion matrix, without normalization')

#     print(cm)
    plt.figure(figsize=(10, 10))
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()


def data_augmentation(data, length=400, dimension=183):
    augmented_data_tmp = np.zeros([length, dimension])
    augmented_data = np.zeros([0, dimension])
    for i in range(int(np.max(data[:, 0])) + 1):
        index = data[:, 0] == i
        data_tmp = data[index, :]
        std = np.std(data_tmp, axis=0)
        mean = np.mean(data_tmp, axis=0)
        augmented_data_tmp[:data_tmp.shape[0], :] = data_tmp
        added_data = np.random.normal(
            mean, std, size=(length - data_tmp.shape[0], 183))
        augmented_data_tmp[data_tmp.shape[0]:, ] = added_data
        augmented_data = np.vstack([augmented_data, augmented_data_tmp])
    return augmented_data


def downsampling(data, width=20, stride=20):
    len_bf = data.shape[0]
    len_af = (len_bf - width) / stride + 1
    downsampled_data = np.zeros([int(len_af), data.shape[1]])
    for i in range(int(len_af)):
        downsampled_data[i, :] = np.sum(
            data[i * width:(i + 1) * width, :], axis=0) / width
    return downsampled_data

def scatter_plot(pred,target,label,x_range=[0,6],y_range=[0,6]):
    plt.figure(figsize=(14,8))
    plt.subplot(121)
    plt.xlim(xmin=x_range[0],xmax=x_range[1])
    plt.ylim(ymin=y_range[0],ymax=y_range[1])
    for i in range(int(len(np.unique(label)))):
        plt.scatter(pred[label==i,0],pred[label==i,1])
    plt.subplot(122)
    plt.xlim(xmin=x_range[0],xmax=x_range[1])
    plt.ylim(ymin=y_range[0],ymax=y_range[1])
    for i in range(int(len(np.unique(label)))):
        plt.scatter(target[label==i,0],target[label==i,1])


    
def box_plot(pred,target,label,per_num=100):
    err=np.sqrt(np.sum(np.square(pred-target),axis=1))
    box_test=np.hstack([label.reshape([-1,1]),err.reshape([-1,1])])
    err_tmp=np.zeros([0,per_num])
    label_term=np.unique(label)
    for i in label_term:
        err_tmp=np.vstack([err_tmp,box_test[box_test[:,0]==i,1].reshape([1,-1])])
    df_test=pd.DataFrame(err_tmp.T,columns=[str(i) for i in label_term])
    err_mean=np.array(df_test.describe().T.iloc[:,5])
    for i in label_term:
        err_tmp=np.vstack([err_tmp,box_test[box_test[:,0]==i,1].reshape([1,-1])])
    df_test.boxplot()

def aggregation(pred,width=10):
    aggregation_pred=np.zeros([int(pred.shape[0]/width),pred.shape[1]])
    for i in range(int(pred.shape[0]/width)):
        aggregation_pred[i,:]=np.sum(pred[i*width:(i+1)*width,:],axis=0)/width
    return aggregation_pred

def localization(proba,k=5):
    result_matrix=np.zeros(proba.shape)
    index=np.argsort(proba)[:,-k:]
    for i in range(index.shape[0]):
        for j in range(k):
            result_matrix[i,index[i,j]]=1/k
    return result_matrix
def get_reference_location(train_target,train_label):
    location=np.zeros([int(np.max(train_label))+1,2])
    for i in range(int(np.max(np.unique(train_label)))+1):
        index=train_label==i
        location[i,:]=train_target[index,:][0,:]
    return location
def evalution(result_matrix,reference_location):
    return np.matmul(result_matrix,reference_location)

def sift_location(label,sift_list):
    index=np.array(np.zeros(label.shape[0]),dtype='bool')
    for k in sift_list:
        index=index+(label==k)
    return index

def space_average(pred,label,stride=3):
    label_unique=np.unique(label)
    pred_final=np.zeros(pred.shape)
    for i in label_unique:
        index_tmp=label==i
        pred_tmp=pred[index_tmp,:]
        pred_average=np.zeros(pred_tmp.shape)
        for j in range(pred_tmp.shape[0]):
            if j+1-stride>=0:
                sum_lst=[pred_tmp[j+1-stride+k,:] for k in range(stride)]
                pred_average[j,:]=reduce(lambda x,y:x+y,sum_lst)/len(sum_lst)
            else:
                sum_lst=[pred_tmp[j-j+k,:] for k in range(j+1)]
                pred_average[j,:]=reduce(lambda x,y:x+y,sum_lst)/len(sum_lst)
        pred_final[index_tmp,:]=pred_average
    return pred_final

def KL_divergence(gmm1,gmm2):
    cov1=np.sqrt(gmm1.covariances_[0,:,:])
    cov2=np.sqrt(gmm2.covariances_[0,:,:])
    mean1=gmm1.means_
    mean2=gmm2.means_
    term1=np.log(np.linalg.det(cov1)/np.linalg.det(cov2))
    term2=np.trace(np.matmul(np.linalg.inv(cov1),cov2))-30
    term3=np.matmul(np.matmul(mean2-mean1,np.linalg.inv(cov1)),np.transpose(mean2-mean1))
    term_1=(term1+term2+term3)/2
    
    cov1=np.sqrt(gmm2.covariances_[0,:,:])
    cov2=np.sqrt(gmm1.covariances_[0,:,:])
    mean1=gmm2.means_
    mean2=gmm1.means_
    term1=np.log(np.linalg.det(cov1)/np.linalg.det(cov2))
    term2=np.trace(np.matmul(np.linalg.inv(cov1),cov2))-30
    term3=np.matmul(np.matmul(mean2-mean1,np.linalg.inv(cov1)),np.transpose(mean2-mean1))
    term_2=(term1+term2+term3)/2
    
    return term_1+term_2

def cdf_plot(value,bins=10):
    sorted_data=np.sort(value.flatten())
    percent=[]
    for i in range(sorted_data.shape[0]):
        percent.append(i/sorted_data.shape[0])
#     plt.figure(figsize=(10,5))
#     plt.subplot(121)
#     plt.hist(sorted_data,bins=10)
#     plt.subplot(122)
    plt.plot(sorted_data,percent)

def KL_kernel(train_data,test_data,train_label,test_label,stride=50,start=0,end=6,feature_stride=30):
    KL_matrix=np.zeros([len(np.unique(train_label)),len(np.unique(test_label))])
    j_lst=[]
    i_lst=[]

    for i in np.unique(train_label):
        train_index=train_label==i
        for j in np.unique(test_label):
            test_index=test_label==j
            KL_tmp=0
            for m in range(end-start):
                gmm1=GaussianMixture(n_components=1)
                gmm1.fit(train_data[train_index,m*feature_stride+start*feature_stride:(m+1)*feature_stride+start*feature_stride])
                gmm2=GaussianMixture(n_components=1)
                gmm2.fit(test_data[test_index,m*feature_stride+start*feature_stride:(m+1)*feature_stride+start*feature_stride])
                KL=KL_divergence(gmm1,gmm2)
                KL_tmp=KL_tmp+KL
                KL_matrix[int(i)-1,int(j)-1]=KL_tmp.flatten()
    return KL_matrix

def KL_kernel_new(gmm_matrix_train,gmm_matrix_test):
    KL_matrix=np.zeros([len(gmm_matrix_test)*len(gmm_matrix_test[0]),len(gmm_matrix_train)*len(gmm_matrix_train[0])])
    count1=0
    for i in range(len(gmm_matrix_test)):
        for j in range(len(gmm_matrix_test[0])):
            gmm_test_tmp=gmm_matrix_test[i][j]
            count2=0
            for n in range(len(gmm_matrix_train)):
                for m in range(len(gmm_matrix_train[0])):
                    gmm_train_tmp=gmm_matrix_train[n][m]
                    KL_tmp=0
                    for k in range(len(gmm_matrix_test[0][0])):
                        KL=KL_divergence(gmm_test_tmp[k],gmm_train_tmp[k])
                        KL_tmp=KL_tmp+KL
                    KL_matrix[count1,count2]=KL_tmp
                    count2=count2+1
            count1=count1+1
    return KL_matrix


# def KL_kernel_new(gmm_matrix_train,gmm_matrix_test):
#     KL_matrix=np.zeros([len(gmm_matrix_test)*len(gmm_matrix_test[0]),len(gmm_matrix_train)*len(gmm_matrix_train[0])])
#     count1=0
#     for i in range(len(gmm_matrix_test)):
#         for j in range(len(gmm_matrix_test[0])):
#             gmm_test_tmp=gmm_matrix_test[i][j]
#             count2=0
#             for n in range(len(gmm_matrix_train)):
#                 for m in range(len(gmm_matrix_train[0])):
#                     gmm_train_tmp=gmm_matrix_train[n][m]
#                     KL=KL_divergence(gmm_test_tmp,gmm_train_tmp)
#                     KL_matrix[count1,count2]=KL
#                     count2=count2+1
#             count1=count1+1
#     return KL_matrix


def reshape_data(data,label,target):
    data_reshape=np.zeros([0,data.shape[1]])
    label_reshape=np.zeros([0,1])
    target_reshape=np.zeros([0,2])
    for i in np.unique(label):
        index=label==i
        data_tmp=data[index,:]
        label_tmp=label[index].reshape([-1,1])
        target_tmp=target[index,:]
        data_reshape=np.vstack([data_reshape,data_tmp])
        label_reshape=np.vstack([label_reshape,label_tmp])
        target_reshape=np.vstack([target_reshape,target_tmp])
    return data_reshape,target_reshape,label_reshape


def gmm_production(data,label,stride=100):
    gmm_matrix=[]
    sample_num=data.shape[0]/stride

    for i in np.unique(label):
        print(i)
        index=label==i
        data_tmp=data[index,:]
        num_tmp=int(data_tmp.shape[0]/stride)
        gmm_lst=[]
        for j in range(num_tmp):
            gmm_str=[]
            for m in range(6):
                gmm=GaussianMixture(n_components=1)
                gmm.fit(data_tmp[stride*j:stride*(j+1),30*m:30*(m+1)])
                gmm_str.append(gmm)
            gmm_lst.append(gmm_str)
        gmm_matrix.append(gmm_lst)
    return gmm_matrix

# def gmm_production(data,label,stride=100):
#     gmm_matrix=[]
#     sample_num=data.shape[0]/stride

#     for i in np.unique(label):
#         print(i)
#         index=label==i
#         data_tmp=data[index,:]
#         num_tmp=int(data_tmp.shape[0]/stride)
#         gmm_lst=[]
#         for j in range(num_tmp):
# #             for m in range(6):
#             gmm=GaussianMixture(n_components=1)
#             gmm.fit(data_tmp[stride*j:stride*(j+1),:60])
#             gmm_lst.append(gmm)
#         gmm_matrix.append(gmm_lst)
#     return gmm_matrix
```
